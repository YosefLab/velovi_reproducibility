# RNA velocity in pancreatic endocrinogenesis

RNA velocity analysis with the *EM* and *steady-state models* using data preprocessed with `starsolo`.

**Requires**

* `adata_generation.ipynb`
* `velocyto_var_names.csv` from `velocyto_vi.ipynb`

**Output**

* `DATA_DIR/pancreas/velocities/starsolo_steady_state.pickle`
* `DATA_DIR/pancreas/velocities/starsolo_em.pickle`

## Library imports

```python
import sys

import pandas as pd

import scanpy as sc
import scvelo as scv

sys.path.insert(0, "../../../")
from paths import DATA_DIR
```

```python
sc.logging.print_version_and_date()
```

## General settings

```python
# set verbosity levels
sc.settings.verbosity = 2
scv.settings.verbosity = 3
```

```python
scv.settings.set_figure_params('scvelo', dpi_save=400, dpi=80, transparent=True, fontsize=20, color_map='viridis')
scv.settings.plot_prefix = ""
```

## Constants

```python
N_JOBS = 8
```

```python
VELOCYTO_VAR_NAMES = pd.read_csv('velocyto_var_names.csv', index_col=0, header=None).index.tolist()
```

## Data loading

```python
adata = sc.read(DATA_DIR / 'pancreas' / "starsolo.h5ad")
adata
```

```python
scv.pl.proportions(adata)
```

## Data pre-processing

```python
scv.pp.filter_and_normalize(adata, min_shared_counts=20, n_top_genes=2000, retain_genes=VELOCYTO_VAR_NAMES)
scv.pp.moments(adata, n_pcs=30, n_neighbors=30)
```

## Model fitting

### Steady state

```python
scv.tl.velocity(adata, mode="steady_state")
adata = adata[:, VELOCYTO_VAR_NAMES].copy()
```

```python
pd.DataFrame(
    adata.layers['velocity'],
    index=adata.obs_names,
    columns=adata.var_names
).to_pickle(
    DATA_DIR / 'pancreas' / 'velocities' / 'starsolo_steady_state.pickle'
)
```

### EM

```python
scv.tl.recover_dynamics(adata, var_names='all', n_jobs=N_JOBS)
scv.tl.velocity(adata, mode='dynamical')
```

```python
pd.DataFrame(
    adata[:, ~adata.var['fit_alpha'].isnull()].layers['velocity'],
    index=adata.obs_names,
    columns=adata.var_names[~adata.var['fit_alpha'].isnull()]
).to_pickle(
    DATA_DIR / 'pancreas' / 'velocities' / 'starsolo_em.pickle'
)
```
